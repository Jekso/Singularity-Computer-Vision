# Sort By Area

```python
import cv2
import numpy as np

# Function to calculate all contours area
def get_contour_areas(contours):
    return [cv2.contourArea(cnt) for cnt in contours]


# Find Contours
image = original_image = cv2.imread('../Data/Images/bunchofshapes.jpg')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edged = cv2.Canny(gray, 50, 200)
_, contours, hierarchy = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)


# Contour Areas before sorting
print(get_contour_areas(contours))

# Sorting...
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)

# Contour Areas after sorting
print(get_contour_areas(sorted_contours))


# Draw Contours One By One related to its Area
for c in sorted_contours:
    cv2.drawContours(original_image, [c], -1, (0, 0, 0), 7)
    cv2.waitKey(0)
    cv2.imshow('Contours by area', original_image)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Contours By Area')
    plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Sort By Horizontal Position

```python
import cv2
import numpy as np



# Function to get centroid position
def get_contour_centroid(contour):
    M = cv2.moments(contour)
    cx = int(M['m10'] / M['m00'])
    cy = int(M['m01'] / M['m00'])
    return cx, cy


# Function to get centroid X position
def get_contour_centroid_X_position(contour):
    cx, _ = get_contour_centroid(contour)
    return cx




# Find Contours
image = original_image = cv2.imread('../Data/Images/bunchofshapes.jpg')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edged = cv2.Canny(gray, 50, 200)
_, contours, hierarchy = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)



# Draw Centroids
for (i, c) in enumerate(contours):
    cx, cy = get_contour_centroid(c)
    cv2.circle(original_image,(cx,cy), 10, (0,255,0), -1)
cv2.imshow('Contours Centroids', original_image)

plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Contours Centroids')
plt.show()



# Contour Centroid X Positions
print(get_contour_centroid_X_position(contours[1]))



# Sorting...
sorted_contours = sorted(contours, key=get_contour_centroid_X_position, reverse=False)


# Draw Contours One By One related to its X Position
for c in sorted_contours:
    cv2.drawContours(original_image, [c], -1, (0, 0, 0), 7)
    cv2.waitKey(0)
    cv2.imshow('Contours Centroids', original_image)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Contours Centroids')
    plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
