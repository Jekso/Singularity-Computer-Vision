```python
import pytesseract
import cv2
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'

def process_data(text):
    tokens = text.split("\n\n")
    return tokens[1],tokens[3] 


def read_card(card_img):
    img = cv2.imread('../Data/Images/'+card_img+'.jpg')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    card_num, username = process_data(pytesseract.image_to_string(img))
    print("Card Number is: " + card_num + ' ---- belongs to user: ' + username)
    
for i in range(1, 4):
    read_card('visa' + str(i))
```
