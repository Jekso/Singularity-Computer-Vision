# Import OpenCV Library

```python
import cv2
import matplotlib.pyplot as plt
```

# Read Image

```python
image = cv2.imread('../Data/Images/input.jpg')

cv2.imshow('Image', image)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Hello World')
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

# Get Imge Shape (Width & Height)

```python
image.shape
```

```python
print("Height is: " + str(image.shape[0]) + " Pixel")
print("Width is: " + str(image.shape[1]) + " Pixel")
print("Channels size are : " + str(image.shape[2]) + " Channels")
```

```python
print(image)
```

```python
print(image[10,20])
print(image[100,250])
```

# Write Image

```python
cv2.imwrite('../Data/Output/output.jpg', image)
```
