# Splitting Image Channels

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../Data/Images/input.jpg')
cv2.imshow('Main Image', img)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Main Image')
plt.show()

print(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
b, g, r = cv2.split(img)

cv2.imshow('Blue Channel', b)
cv2.waitKey()

cv2.imshow('Green Channel', g)
cv2.waitKey()

cv2.imshow('Red Channel', r)
cv2.waitKey()

plt.imshow(cv2.cvtColor(b, cv2.COLOR_GRAY2RGB))
plt.title('Blue Channel')
plt.show()

plt.imshow(cv2.cvtColor(g, cv2.COLOR_GRAY2RGB))
plt.title('Green Channel')
plt.show()

plt.imshow(cv2.cvtColor(r, cv2.COLOR_GRAY2RGB))
plt.title('Red Channel')
plt.show()

print(b)
# print(g)
# print(r)


cv2.destroyAllWindows()
```

# Merge Channels

```python
img = cv2.merge([b,g,r])

cv2.imshow('Merged', img)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Merged')
plt.show()

print(img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.merge([b+200, g, r])
cv2.imshow('Merged with increase blue', img)
cv2.waitKey()

img2 = cv2.merge([b, g+200, r])
cv2.imshow('Merged with increase green', img2)
cv2.waitKey()

img3 = cv2.merge([b, g, r+200])
cv2.imshow('Merged with increase red', img3)
cv2.waitKey()


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Merged with increase blue')
plt.show()

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Merged with increase green')
plt.show()

plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.title('Merged with increase red')
plt.show()

print(img)

cv2.destroyAllWindows()
```

# Show each Channel

```python
import cv2
import numpy as np

B, G, R = cv2.split(img)
 
zeros = np.zeros(img.shape[:2], dtype = "uint8")

cv2.imshow("Red", cv2.merge([zeros, zeros, R]))
cv2.waitKey(0)

cv2.imshow("Green", cv2.merge([zeros, G, zeros]))
cv2.waitKey(0)

cv2.imshow("Blue", cv2.merge([B, zeros, zeros]))
cv2.waitKey(0)

plt.imshow(cv2.cvtColor(cv2.merge([zeros, zeros, R]), cv2.COLOR_BGR2RGB))
plt.title('Red')
plt.show()

plt.imshow(cv2.cvtColor(cv2.merge([zeros, G, zeros]), cv2.COLOR_BGR2RGB))
plt.title('Green')
plt.show()

plt.imshow(cv2.cvtColor(cv2.merge([B, zeros, zeros]), cv2.COLOR_BGR2RGB))
plt.title('Blue')
plt.show()

cv2.destroyAllWindows()
```
