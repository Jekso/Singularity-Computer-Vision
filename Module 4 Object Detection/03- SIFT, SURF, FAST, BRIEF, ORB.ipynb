# SIFT

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../Data/Images/input.jpg', 0)

sift = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(image, None)

print("Number of keypoints Detected: ", len(keypoints))

img2 = cv2.drawKeypoints(image, keypoints, color=(0,255,0), outImage=image, flags=cv2.DRAW_MATCHES_FLAGS_DEFAULT)

cv2.imshow('Feature Method - SIFT', img2)
cv2.waitKey()
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Feature Method - SIFT')
plt.show()

cv2.destroyAllWindows()
```

# FAST

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../Data/Images/input.jpg',0)

fast = cv2.FastFeatureDetector_create()

kp = fast.detect(img,None)
img = cv2.drawKeypoints(img, kp, color=(0,255,0), outImage=img)

cv2.imshow('Feature Method - FAST', img)
cv2.waitKey()

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Feature Method - FAST')
plt.show()

cv2.destroyAllWindows()
```

# BRIEF

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../Data/Images/input.jpg',0)

star = cv2.xfeatures2d.StarDetector_create()
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()

kp = star.detect(img,None)
kp, des = brief.compute(img, kp)

img = cv2.drawKeypoints(img, kp, color=(0,255,0), outImage=img)

cv2.imshow('Feature Method - FAST', img)
cv2.waitKey()

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Feature Method - FAST')
plt.show()

cv2.destroyAllWindows()
```
